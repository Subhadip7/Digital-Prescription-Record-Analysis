```python
import pandas as pd
import pymysql
import numpy as np

# ===============================
# DATABASE CONFIG
# ===============================
DB_HOST = "localhost"
DB_PORT = 3306
DB_USER = "root"
DB_PASS = "Subha@@5092"
DB_NAME = "digital_prescription_db"
BATCH_SIZE = 1000

# ===============================
# LOAD CSV FILES
# ===============================
patients = pd.read_csv('patients_1000.csv')
doctors = pd.read_csv('doctors_1000.csv')
medicines = pd.read_csv('medicines_1000.csv')
prescriptions = pd.read_csv('prescriptions_1000.csv')

# ===============================
# CLEAN PATIENTS
# ===============================
patients = patients.drop_duplicates().copy()

patients['patient_id'] = patients['patient_id'].astype(str).str.strip()
patients['patient_name'] = patients['patient_name'].astype(str).str.strip()

# Normalize gender safely
patients['gender'] = patients['gender'].astype(str).str.upper().str[0]
patients['gender'] = patients['gender'].apply(lambda x: x if x in ['M','F'] else 'U')

# Clean age
patients['age'] = pd.to_numeric(patients['age'], errors='coerce')
patients = patients[(patients['age'] >= 0) & (patients['age'] <= 100)]
patients['age'] = patients['age'].astype(int)

# ===============================
# CLEAN DOCTORS
# ===============================
doctors = doctors.drop_duplicates().copy()
doctors['doctor_id'] = doctors['doctor_id'].astype(str).str.strip()
doctors['doctor_name'] = doctors['doctor_name'].astype(str).str.strip()

# ===============================
# CLEAN MEDICINES
# ===============================
medicines = medicines.drop_duplicates().copy()
medicines['medicine_id'] = medicines['medicine_id'].astype(str).str.strip()
medicines['medicine_name'] = medicines['medicine_name'].astype(str).str.strip()

# ===============================
# CLEAN PRESCRIPTIONS
# ===============================
prescriptions = prescriptions.drop_duplicates(
    subset=['prescription_id'], keep=False
).copy()

prescriptions['prescription_id'] = prescriptions['prescription_id'].astype(str).str.strip()
prescriptions['patient_id'] = prescriptions['patient_id'].astype(str).str.strip()
prescriptions['doctor_id'] = prescriptions['doctor_id'].astype(str).str.strip()
prescriptions['medicine_id'] = prescriptions['medicine_id'].astype(str).str.strip()

prescriptions['dosage'] = prescriptions['dosage'].fillna('Not Provided')
prescriptions['frequency'] = prescriptions['frequency'].fillna('Not Provided')

# Convert prescribed_date safely
prescriptions['prescribed_date'] = pd.to_datetime(
    prescriptions['prescribed_date'], errors='coerce'
).dt.date

prescriptions = prescriptions[prescriptions['prescribed_date'].notna()]

# ===============================
# FOREIGN KEY VALIDATION
# ===============================
valid_patients = set(patients['patient_id'])
valid_doctors = set(doctors['doctor_id'])
valid_medicines = set(medicines['medicine_id'])

prescriptions = prescriptions[
    prescriptions['patient_id'].isin(valid_patients) &
    prescriptions['doctor_id'].isin(valid_doctors) &
    prescriptions['medicine_id'].isin(valid_medicines)
]

# ===============================
# SAFE NaN â†’ None CONVERSION
# ===============================
def df_to_rows(df, columns):
    temp = df[columns].copy()
    temp = temp.replace({np.nan: None})
    temp = temp.where(pd.notnull(temp), None)
    return temp.values.tolist()

patient_rows = df_to_rows(
    patients,
    ['patient_id','patient_name','gender','age','contact_no']
)

doctor_rows = df_to_rows(
    doctors,
    ['doctor_id','doctor_name','specialization','hospital_name']
)

medicine_rows = df_to_rows(
    medicines,
    ['medicine_id','medicine_name','category']
)

prescription_rows = df_to_rows(
    prescriptions,
    ['prescription_id','patient_id','doctor_id',
     'medicine_id','dosage','frequency','prescribed_date']
)

# ===============================
# CONNECT TO MYSQL
# ===============================
conn = pymysql.connect(
    host=DB_HOST,
    port=DB_PORT,
    user=DB_USER,
    password=DB_PASS,
    database=DB_NAME,
    autocommit=False
)

cursor = conn.cursor()

def batch_insert(query, rows):
    for i in range(0, len(rows), BATCH_SIZE):
        batch = rows[i:i+BATCH_SIZE]
        cursor.executemany(query, batch)

# ===============================
# INSERT QUERIES
# ===============================

patient_query = """
INSERT INTO patients (patient_id, patient_name, gender, age, contact_no)
VALUES (%s, %s, %s, %s, %s)
ON DUPLICATE KEY UPDATE
patient_name=VALUES(patient_name),
gender=VALUES(gender),
age=VALUES(age),
contact_no=VALUES(contact_no)
"""

doctor_query = """
INSERT INTO doctors (doctor_id, doctor_name, specialization, hospital_name)
VALUES (%s, %s, %s, %s)
ON DUPLICATE KEY UPDATE
doctor_name=VALUES(doctor_name),
specialization=VALUES(specialization),
hospital_name=VALUES(hospital_name)
"""

medicine_query = """
INSERT INTO medicines (medicine_id, medicine_name, category)
VALUES (%s, %s, %s)
ON DUPLICATE KEY UPDATE
medicine_name=VALUES(medicine_name),
category=VALUES(category)
"""

prescription_query = """
INSERT INTO prescriptions
(prescription_id, patient_id, doctor_id, medicine_id,
 dosage, frequency, prescribed_date)
VALUES (%s, %s, %s, %s, %s, %s, %s)
ON DUPLICATE KEY UPDATE
dosage=VALUES(dosage),
frequency=VALUES(frequency),
prescribed_date=VALUES(prescribed_date)
"""

# ===============================
# EXECUTE INSERTS
# ===============================
try:
    batch_insert(patient_query, patient_rows)
    batch_insert(doctor_query, doctor_rows)
    batch_insert(medicine_query, medicine_rows)
    batch_insert(prescription_query, prescription_rows)

    conn.commit()
    print("âœ… Data inserted successfully into MySQL!")

except Exception as e:
    conn.rollback()
    print("âŒ Error:", e)

finally:
    cursor.close()
    conn.close()
```

```python
import pymysql
import pandas as pd
import matplotlib.pyplot as plt

# ===============================
# DATABASE CONFIG
# ===============================
DB_HOST = "localhost"
DB_PORT = 3306
DB_USER = "root"
DB_PASS = "Subha@@5092"
DB_NAME = "digital_prescription_db"

# ===============================
# CONNECT TO MYSQL
# ===============================
conn = pymysql.connect(
    host=DB_HOST,
    port=DB_PORT,
    user=DB_USER,
    password=DB_PASS,
    database=DB_NAME
)

# ===============================
# 1ï¸âƒ£ Most Frequently Prescribed Medicines
# ===============================
query1 = """
SELECT m.medicine_name, COUNT(*) AS total_prescriptions
FROM prescriptions p
JOIN medicines m ON p.medicine_id = m.medicine_id
GROUP BY m.medicine_name
ORDER BY total_prescriptions DESC
LIMIT 10;
"""
df1 = pd.read_sql(query1, conn)

plt.figure()
plt.bar(df1['medicine_name'], df1['total_prescriptions'])
plt.xticks(rotation=45)
plt.title("Top 10 Most Prescribed Medicines")
plt.xlabel("Medicine")
plt.ylabel("Total Prescriptions")
plt.tight_layout()
plt.show()

# ===============================
# 2ï¸âƒ£ Doctor-wise Prescription Volume
# ===============================
query2 = """
SELECT d.doctor_name, COUNT(*) AS total_prescriptions
FROM prescriptions p
JOIN doctors d ON p.doctor_id = d.doctor_id
GROUP BY d.doctor_name
ORDER BY total_prescriptions DESC
LIMIT 10;
"""

df2 = pd.read_sql(query2, conn)

plt.figure()
plt.bar(df2['doctor_name'], df2['total_prescriptions'])
plt.xticks(rotation=45)
plt.title("Top 10 Doctors by Prescription Volume")
plt.xlabel("Doctor")
plt.ylabel("Total Prescriptions")
plt.tight_layout()
plt.show()

# ===============================
# 3ï¸âƒ£ Gender-wise Prescription Distribution
# ===============================
query3 = """
SELECT pt.gender, COUNT(*) AS total_prescriptions
FROM prescriptions p
JOIN patients pt ON p.patient_id = pt.patient_id
GROUP BY pt.gender;
"""
df3 = pd.read_sql(query3, conn)

plt.figure()
plt.bar(df3['gender'], df3['total_prescriptions'])
plt.title("Gender-wise Prescription Distribution")
plt.xlabel("Gender")
plt.ylabel("Total Prescriptions")
plt.tight_layout()
plt.show()

# ===============================
# 4ï¸âƒ£ Age Group vs Number of Prescriptions
# ===============================
query4 = """
SELECT 
    CASE
        WHEN pt.age BETWEEN 0 AND 18 THEN '0-18'
        WHEN pt.age BETWEEN 19 AND 35 THEN '19-35'
        WHEN pt.age BETWEEN 36 AND 60 THEN '36-60'
        ELSE '60+'
    END AS age_group,
    COUNT(*) AS total_prescriptions
FROM prescriptions p
JOIN patients pt ON p.patient_id = pt.patient_id
GROUP BY age_group
ORDER BY age_group;
"""
df4 = pd.read_sql(query4, conn)

plt.figure()
plt.bar(df4['age_group'], df4['total_prescriptions'])
plt.title("Age Group vs Number of Prescriptions")
plt.xlabel("Age Group")
plt.ylabel("Total Prescriptions")
plt.tight_layout()
plt.show()

# ===============================
# 5ï¸âƒ£ Detect Doctors with Unusually High Prescriptions
# ===============================
query5 = """
SELECT d.doctor_name, COUNT(*) AS total_prescriptions
FROM prescriptions p
JOIN doctors d ON p.doctor_id = d.doctor_id
GROUP BY d.doctor_name
ORDER BY total_prescriptions DESC;
"""
df5 = pd.read_sql(query5, conn)

mean_val = df5['total_prescriptions'].mean()
std_val = df5['total_prescriptions'].std()

df5['high_flag'] = df5['total_prescriptions'] > (mean_val + 2 * std_val)

plt.figure()
plt.bar(df5['doctor_name'], df5['total_prescriptions'])
plt.xticks(rotation=90)
plt.title("Doctor Prescription Outlier Detection")
plt.xlabel("Doctor")
plt.ylabel("Total Prescriptions")
plt.tight_layout()
plt.show()

# ===============================
# 6ï¸âƒ£ Daily Prescription Trend
# ===============================
query6 = """
SELECT prescribed_date, COUNT(*) AS total_prescriptions
FROM prescriptions
GROUP BY prescribed_date
ORDER BY prescribed_date;
"""
df6 = pd.read_sql(query6, conn)

plt.figure()
plt.plot(df6['prescribed_date'], df6['total_prescriptions'])
plt.xticks(rotation=45)
plt.title("Daily Prescription Trend")
plt.xlabel("Date")
plt.ylabel("Total Prescriptions")
plt.tight_layout()
plt.show()

# ===============================
# 7ï¸âƒ£ Data Quality Comparison (Before vs After Cleaning)
# ===============================
# Assuming you still have original CSV row counts
raw_patients = pd.read_csv('patients_1000.csv')
raw_prescriptions = pd.read_csv('prescriptions_1000.csv')

query_clean_counts = """
SELECT 
(SELECT COUNT(*) FROM patients) AS clean_patients,
(SELECT COUNT(*) FROM prescriptions) AS clean_prescriptions;
"""
df_clean = pd.read_sql(query_clean_counts, conn)

labels = ['Patients', 'Prescriptions']
raw_counts = [len(raw_patients), len(raw_prescriptions)]
clean_counts = [df_clean['clean_patients'][0], df_clean['clean_prescriptions'][0]]

x = range(len(labels))

plt.figure()
plt.bar(x, raw_counts)
plt.xticks(x, labels)
plt.title("Raw Data Volume")
plt.ylabel("Row Count")
plt.tight_layout()
plt.show()

plt.figure()
plt.bar(x, clean_counts)
plt.xticks(x, labels)
plt.title("Cleaned Data Volume")
plt.ylabel("Row Count")
plt.tight_layout()
plt.show()

# ===============================
# CLOSE CONNECTION
# ===============================
conn.close()

print("âœ… All SQL analytics and dashboards generated successfully!")
```
